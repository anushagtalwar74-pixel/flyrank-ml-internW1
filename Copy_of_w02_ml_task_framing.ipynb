## 1. My lane as an ML task (type)

## My lane as an ML task (type)

My chosen lane is Content Refresh Prioritization.

This is a **classification** problem because the goal is to classify each content page as likely to decline or not decline based on historical search performance data. The output helps identify which pages should be prioritized for content refresh.

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

## Target or proxy

The target is **is_declining_label**.

This label is created from the observed value of **trend_direction**. A page is labeled as declining when trend_direction is "down". The model predicts this label using other page performance features.

## 3. Success metric

*One metric you can defend. What number means 'good'?*

## Success metric

The success metric is **Precision@50**.

Precision@50 measures how many of the top 50 pages selected by the model are actually declining. A higher Precision@50 means the model is identifying the most important pages for refresh more accurately.

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

## Why ML beats a fixed rule here

A fixed rule may consider only one or two features, such as impressions or CTR.

Machine Learning can learn patterns by combining many features like impressions, CTR, average position, content age, and engagement metrics. Different pages decline for different reasons, so ML can detect relationships that are difficult to capture using simple if-else rules.

The model helps content teams make better decisions about which pages should be refreshed first.
